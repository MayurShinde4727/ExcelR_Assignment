```python
import pandas as pd
import numpy as np
import scipy
import scipy as stats
import statsmodels.api as sm
from scipy.stats import norm
```

# Question 1


```python
data = pd.read_csv('Cutlets.csv')
data.head()
```

```python
unitA = pd.Series(data['Unit A'])
unitA
```

```python
unitB = pd.Series(data['Unit B'])
unitB
```

```python
# It is a 2 sample t-test
# Null hypothesis = mean1 = mean2
# Alternate hypothesis = mean1 != mean2
sm.stats.ttest_ind(unitA,unitB)
```

pvalue = 0.47 which is more than 0.05, so we fail to reject the null hypothesis

```python
data3 = pd.read_csv('BuyerRatio.csv')
data3
```

```python
males = [50,142,131,70]
females= [435,1523,1356,750]
sales = [males,females]
sales
```

```python
scipy.stats.chi2_contingency(sales)
```

pvalue = 0.6603, since pvalue is greater than 0.05.
so we fail to reject the null hypothesis.
